```python
metric_str = 'Prob<3'
colorbar_label = 'Prob < 3 $\AA$'
for pdb in visualize_sadlsa_results.query_score_df(score_df)[:,0]:
    print(pdb)
    pdbid = pdb.split('_')[0]
    chainid = pdb.split('_')[1]
    # collecting metric values for specific pdb id and chain
    temp = visualize_sadlsa_results.query_alignment_df(align_df, pdb,metric_str)
    # collecting structure object
    structure_file = visualize_sadlsa_results.grab_structure(pdbid)
    # updating the b-factor column of the structure for visualization purposes; saving a pdb file of the chain of interest
    structure_file = visualize_sadlsa_results.edit_pdb(structure_file,pdb,temp,metric_str,working_dir=working_dir)
    if structure_file:
        # creating VMD vis state of the structure of interest
        visualize_sadlsa_results.create_vmd_vis_state(working_dir+'%s_%s.vmd'%(pdb,metric_str),
                         working_dir+'%s_%s_colorbar.png'%(pdb,metric_str),
                         working_dir+'%s_%s.pdb'%(pdb,metric_str),1.0,'blue',colorbar_label=colorbar_label)
```
