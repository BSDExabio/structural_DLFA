### VMD Visualization of top hits

```python
metric_str = 'Prob<3'
colorbar_label = 'Prob < 3 $\AA$'
for pdb in visualize_sadlsa_results.query_score_df(score_df)[:,0]:
    pdbid = pdb.split('_')[0]
    chainid = pdb.split('_')[1]
    # collecting metric values for specific pdb id and chain
    temp = visualize_sadlsa_results.query_alignment_df(align_df, pdb,metric_str)
    # making residue indices zero-indexed rather than 1 to avoid weirdness
    temp[:,0] = edit_pdbs.make_zero_indexed(temp[:,0])
    # collecting structure object
    structure_file = edit_pdbs.grab_structure(pdbid)
    uniprot_id = sadlsa_parsers._get_uniprot(pdbid,chainid,structure_file)
    print(pdb,uniprot_id)
    # updating the b-factor column of the structure for visualization purposes; saving a pdb file of the chain of interest
    structure_file = edit_pdbs.edit_pdb(structure_file, temp, pdb+'_'+metric_str, working_dir=working_dir, sel_string = "protein and segid %s"%(chainid))
    if structure_file:
        # creating VMD vis state of the structure of interest
        visualize_sadlsa_results.create_vmd_vis_state(working_dir+'%s_%s.vmd'%(pdb,metric_str),
                         working_dir+'%s_%s_colorbar.png'%(pdb,metric_str),
                         working_dir+'%s_%s.pdb'%(pdb,metric_str),1.0,'blue',colorbar_label=colorbar_label)
```

### NGLView Visualization of top hits

```python
import nglview as nv
from nglview.color import ColormakerRegistry
print("nglview version = {}".format(nv.__version__))
```

```python
metric_str = 'Prob<3'
colorbar_label = 'test'
colorbar_color = 'green'

pdb = visualize_sadlsa_results.query_score_df(score_df)[0,0]
pdbid = pdb.split('_')[0]
chainid = pdb.split('_')[1]

temp = visualize_sadlsa_results.query_alignment_df(align_df, pdb, metric_str)
temp[:,0] = edit_pdbs.make_zero_indexed(temp[:,0])
print(temp[:10])

structure_file = edit_pdbs.grab_structure(pdbid)

import MDAnalysis
import numpy as np
import warnings

# loading the mmtf structure object into an MDAnalysis universe object
u = MDAnalysis.Universe(structure_file)
# creating the atom selection groups for the structure
sel = edit_pdbs.make_atm_sel(u,"protein")
#_all = make_atm_sel(u,'all')

# moving all atoms of chain of interest to have the CoM to be the origin
#_all.translate(-sel.center_of_mass())
sel.translate(-sel.center_of_mass())
    
# creating the one-indexed array of sel's residue indices of resolved 
# residues; used to mirror the one-indexed SAdLSA resid column
nRes_range = range(sel.n_residues)
for i in nRes_range:
    if i in temp[:,0]:
        idx = np.argwhere(i == temp[:,0])
        sel.residues[i].atoms.tempfactors = float(temp[idx,1])   # will be printed with 2 decimal points; 
    else:
        sel.residues[i].atoms.tempfactors = 0.0
```

```python
ColormakerRegistry.add_selection_scheme("awesome",[['blue', '0-10']])


colors = [c.replace('#','0x') for c in ['#ffffff','#80c080','#008000']]

b = nv.show_mdanalysis(sel)
b.clear()
b.add_cartoon('protein',color='bfactor')
b
```
